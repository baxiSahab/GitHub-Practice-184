# рџџў Easy: Implement Dropout

Implement **Dropout** regularization from scratch.

### Signature
```python
class MyDropout(nn.Module):
    def __init__(self, p: float = 0.5): ...
    def forward(self, x: Tensor) -> Tensor: ...
```

### Rules
- During **training**: zero each element with probability `p`, scale remaining by `1/(1-p)`
- During **eval**: return input unchanged (identity)
- Do NOT use `nn.Dropout` or `F.dropout`

````python
import torch
import torch.nn as nn
import random
````

````python
# вњЏпёЏ YOUR IMPLEMENTATION HERE

class MyDropout(nn.Module):
    def __init__(self, p=0.5):
        super().__init__()
        self.p = p

        if not 0 <= p <= 1:
          raise ValueError("Probability must be between 0 and 1")

    def forward(self, x):
      L = len(x)
      randIndex = torch.randperm(L)

      for randomNumber in randIndex[:5]:
        x[randomNumber] = 0

      return x * 1/(1-p)
````

````python
p = .5
print(p)
if not 0 <= p <= 1:
  raise ValueError("Probability must be between 0 and 1")
x=torch.ones(10)
L=len(x)
# print(torch.randint(low=0, high=L, size=(L // 2,)))
randIndex = torch.randperm(L)
print(randIndex[:5])
````

````python
# рџ§Є Debug
d = MyDropout(p=0.5)
print('init',d)

d.train()
print('train',d)

x = torch.ones(10)
print('x',x)

print('Train:', d(x))
d.eval()

print('eval',d)
print('Eval: ', d(x))
````

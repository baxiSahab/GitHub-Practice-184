# рџџў Easy: Implement Dropout

Implement **Dropout** regularization from scratch.

### Signature
```python
class MyDropout(nn.Module):
    def __init__(self, p: float = 0.5): ...
    def forward(self, x: Tensor) -> Tensor: ...
```

### Rules
- During **training**: zero each element with probability `p`, scale remaining by `1/(1-p)`
- During **eval**: return input unchanged (identity)
- Do NOT use `nn.Dropout` or `F.dropout`

````python
import torch
import torch.nn as nn
import random
````

````python
# вњЏпёЏ YOUR IMPLEMENTATION HERE

class MyDropout(nn.Module):
    def __init__(self, p=0.5):
        super().__init__()
        self.p = p

        if not 0 <= p <= 1:
          raise ValueError("Probability must be between 0 and 1")

    def forward(self, x):
      if not self.training or self.p == 0:
            return x

      mask = torch.rand_like(x)
      new_x = torch.where(mask < self.p, 0.0, x)

      return new_x * 1/(1-self.p)
````

````python
p = .5
print(p)
if not 0 <= p <= 1:
  raise ValueError("Probability must be between 0 and 1")
x=torch.ones(10)
mask_noise = torch.rand_like(x)
print(mask_noise)
````

````python
# рџ§Є Debug
d = MyDropout(p=0.5)
print('init',d)

d.train()
print('train',d)

x = torch.ones(10)
print('x',x)

print('Train:', d(x))
d.eval()

print('eval',d)
print('Eval: ', d(x))
````

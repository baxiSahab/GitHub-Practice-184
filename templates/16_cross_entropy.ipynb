# 🟢 Easy: Cross-Entropy Loss

Implement **cross-entropy loss** from scratch.

$$\text{CE}(x, y) = -\log\frac{e^{x_y}}{\sum_j e^{x_j}}$$

### Signature
```python
def cross_entropy_loss(logits: Tensor, targets: Tensor) -> Tensor:
    # logits: (B, C) float, targets: (B,) long indices
    # Returns: scalar loss (mean over batch)
```

### Rules
- Do NOT use `F.cross_entropy` or `nn.CrossEntropyLoss`
- Must be numerically stable (use logsumexp trick)

````python
# ✏️ YOUR IMPLEMENTATION HERE

def cross_entropy_loss(logits, targets):

  max_logit = logits.max(keepdim=True, dim=1).values

  sum_exp = torch.exp( logits[ torch.arange(logits.shape(0))  , targets ]  - max_logit ).sum(dim-1, keepdim=True)

  log_sum_exp = torch.log(sum_exp) # + max_logit??

  return torch.log( torch.exp( max_logit - logits[ torch.arange(logits.shape(0))  , targets ] ) )  - logits[torch.arange(logits.shape(0))  , targets]

  # torch.log( torch.exp(logits).sum() ) - logits[targets]

  # -torch.log(torch.exp(max - logits[targets]) / sum).mean()
````

````python
# Demo
logits = torch.randn(4, 10)
targets = torch.randint(0, 10, (4,))
print('Loss:', cross_entropy_loss(logits, targets).item())
print('Ref: ', torch.nn.functional.cross_entropy(logits, targets).item())
````
